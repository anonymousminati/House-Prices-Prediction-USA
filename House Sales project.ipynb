```python

df = df.drop(columns=['date','yr_built','yr_renovated','id','zipcode'],inplace=True)
```

```python

df['price'] = df['price'].astype(int)
```

```python
sns.scatterplot(data=df,y='price',x='sqft_living')
plt.show()
```

```python
sns.histplot(data=df,x='price')
plt.show()
```

```python
df['sqft_living']=df[['sqft_living']].astype(int)
df['sqft_lot']=df[['sqft_lot']].astype(int)
df['sqft_above']=df[['sqft_above']].astype(int)
```

```python
df = df.drop(columns=['date'])
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, cmap='coolwarm')
```

**Machine Learning**

```python
X = df.drop(columns=['price','sqft_living'])
y = df['price']
```

**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=45)
```

```python
len(X_train), len(X_test), len(y_train),len(y_test)
```

```python
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
test_2 = mean_squared_error(y_test,prediction)
test_3 = r2_score(y_test,prediction)
```

```python
print("Mean Squared Error:", test_2)
print("R squared Error:", test_3)
```

**OLS Model**

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y,X).fit()  
print(model.summary())
```
